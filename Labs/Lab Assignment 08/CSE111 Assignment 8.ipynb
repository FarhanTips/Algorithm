```python
#Task 1
class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())
class ComplexNumber(RealNumber):
    def __init__(self,r=1,im=1):
        super().__init__(r)
        self.im=im
    def __str__(self):
        return f"{super().__str__()}.0\nImaginaryPart: {self.im}.0"
cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
#Task2
class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
class ComplexNumber(RealNumber):
    def __init__(self,number,u):
        super().__init__(number)
        self.u=u
    def __str__(self):
        if self.u>=0:
            return super().__str__()+" + "+str(self.u)+"i"
        else:
            return super().__str__()  +" "+ str(self.u)[0:1]+" "+ str(self.u)[1:]+ "i"
    def __add__(self, other):
        s1=self.number + other.number.number  #it means r1 is a memory location and when r1 is called then 'other.number' is
#calling r1 but r1 is a location or object. Then we need 'r1.number' means 'other.number.number' which will give us the value of r1.
        s2=self.u+other.u
        return ComplexNumber(s1,s2)
    def __sub__(self, other):
        s1=self.number - other.number.number   
        
        s2=self.u-other.u
        return ComplexNumber(s1,s2)
r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

**OUTPUT:**

**----------------------------------**

```python
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self,marks,time,sub1,sub2,sub3=None):
        super().__init__(marks)
        self.time=time
        self.sub1=sub1
        self.sub2=sub2
        self.sub3=sub3
    def __str__(self):
        if self.sub3==None:
            return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: 4"
        else:
            return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: 5"
    def examSyllabus(self):
        if self.sub3==None:
            return f"{super().examSyllabus()} , {self.sub1} , {self.sub2}"
        else:
            return f"{super().examSyllabus()} , {self.sub1} , {self.sub2} , {self.sub3}"
    def examParts(self):
        if self.sub3==None:
            return f"{super().examParts()}Part 3 - {self.sub1}\nPart 4 - {self.sub2}"
        else:
            return f"{super().examParts()}Part 3 - {self.sub1}\nPart 4 - {self.sub2}\nPart 5 - {self.sub3}"
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture =ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

**OUTPUT:** 

------------Basic Info:--------------

```python
#Task 7
class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0,weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type
    def get_type(self):
        return 'Main type: ' + self.type
    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'
    def __str__(self):
        return "Name: " + self.name + ", HP: " +str(self.hit_point) + ", Weakness: " + self.weakness
class PokemonExtra(PokemonBasic):
    def __init__(self,name , hp,weakness, type,sectype=None,otmove=None):
        super().__init__(name , hp,weakness, type)
        self.sectype=sectype
        self.otmove=otmove

    def get_type(self):
        if self.sectype==None:
            return f"{super().get_type()}"
        else:
            return f"{super().get_type()}, Secondary type: {self.sectype}"
    def get_move(self):
        if self.otmove!=None:
            s=""
            for i in self.otmove:
                s+=i+", "
            s=s+"\b\b"
            return f"{super().get_move()}\nOther Move: {s}"

        else:
            return f"{super().get_move()}"
print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water','Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water',
'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```
